```python
def cek_kelulusan(nilai_ulangan):
    for nilai in nilai_ulangan:
        if nilai >= 70:
            print(f"Nilai {nilai} lulus.")
        else:
            print(f"Nilai {nilai} tidak lulus.")

# Memasukkan beberapa nilai ulangan untuk diuji
nilai_ulangan_uji = [65, 80, 45, 90, 60]

# Memanggil fungsi untuk menguji nilai ulangan
cek_kelulusan(nilai_ulangan_uji)
```

```python
def cek_kelulusan(nilai_ulangan):
    i = 0
    while i < len(nilai_ulangan):
        nilai = nilai_ulangan[i]
        if nilai >= 70:
            print(f"Nilai {nilai} lulus.")
        else:
            print(f"Nilai {nilai} tidak lulus.")
        i += 1

# Memasukkan beberapa nilai ulangan untuk diuji
nilai_ulangan_uji = [65, 80, 45, 90, 60]

# Memanggil fungsi untuk menguji nilai ulangan
cek_kelulusan(nilai_ulangan_uji)
```
